## **Assignment 1**

1. Load the `CSV` manually using the csv module (not pandas).
2. Store the consumption column in a list of floats.
3. Compute:

- 3.1 *Total consumption*

- 3.2 *Average consumption*

- 3.3 *Maximum and minimum*

4. Use a list comprehension to convert hourly consumption into kWh.
5. Print all results in a clean format.

```python
import csv

```

```python


consumption = []

with open(r'C:\Users\Muhib Ur-Rahman\Desktop\intro to ai\Assignment no 1.csv', 'r') as csvfile:
    reader = csv.DictReader(csvfile) 
    for row in reader:
        print(row)
        consumption.append(int(row['consumption(w)']))


```

```python

# 2. Computations
total_consumption = sum(consumption)
average_consumption = total_consumption / len(consumption)
max_consumption = max(consumption)
min_consumption = min(consumption)
```

```python
print("Total consumption :", total_consumption, "Wh")
print("Average consumption :", average_consumption, "Wh")
print("Maximum consumption :", max_consumption, "Wh")
print("Minimum consumption :", min_consumption, "Wh")
```

```python
# 3. Convert to kWh using list comprehension
consumption_kwh = [w / 1000 for w in consumption]
print(f"list of consumption in kwh: {consumption_kwh}")
```
