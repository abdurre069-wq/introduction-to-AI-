###	**Group consumption by hour (manual group-by)**
Create a dictionary like:

`{ 1: [200, 700, 800, 500],`
  
  `2: [300, 600, 700, 400],`
  
  `3: [500, 400, 600, 650] }`

`2.1.	Compute average consumption per hour`

`2.2.	Compute peak consumption per hour`

`2.3.	Print results in a neat format`

```python
consumption_by_hours={ 1: [200, 700, 800, 500],
  
  2: [300, 600, 700, 400],
  
  3: [500, 400, 600, 650] }
```

```python
for hour, values in consumption_by_hours.items():
    avg = sum(values) / len(values)
    print(f"Average consumption of hour {hour} is: {avg}")

```

```python
for hour, values in consumption_by_hours.items():
    peak_cons=max(values)
    print(f"peak consumption of hour {hour} is: {peak_cons}")
```
